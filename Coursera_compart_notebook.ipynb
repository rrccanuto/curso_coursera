# Principais bibliotecas de Data Science em Python

1. **NumPy**  
   - Base para cálculos numéricos e matrizes multidimensionais.  
   - Altamente otimizada para desempenho matemático.  

2. **Pandas**  
   - Especializada em manipulação e análise de dados em tabelas (DataFrames).  
   - Ideal para limpeza, transformação e exploração de dados.  

3. **Matplotlib**  
   - Biblioteca de visualização básica e poderosa.  
   - Cria gráficos 2D de alta qualidade para relatórios e dashboards.  

4. **Seaborn**  
   - Construída sobre o Matplotlib, mas com foco em visualizações estatísticas elegantes.  
   - Facilita análises exploratórias rápidas e visualmente agradáveis.  

5. **Scikit-learn**  
   - Ferramenta principal para *Machine Learning* tradicional (regressão, classificação, clustering).  
   - Inclui métricas, pipelines e algoritmos prontos para uso.

### Abaixo estão alguns exemplos de avaliação de expressões aritméticas em Python

| **Expressão** | **Descrição** | **Resultado Esperado** |
|------------|------------|--------------------|
| `5 + 3` | Soma de dois números inteiros | `8` |
| `10 - 4` | Subtração | `6` |
| `7 * 2` | Multiplicação | `14` |
| `8 / 2` | Divisão comum (retorna float) | `4.0` |
| `9 // 2` | Divisão inteira (descarta o decimal) | `4` |
| `9 % 2` | Módulo — resto da divisão | `1` |
| `2 ** 3` | Exponenciação (2 elevado a 3) | `8` |
| `(5 + 3) * 2` | Uso de parênteses para alterar a precedência | `16` |
